The subscripts are indices of a matrix or a vector, while superscripts indicate which layer a particular element belongs to. For example, in $w^{(1)}_{21}$, the superscript indicates that the weight is for a neuron in Layer 1, and the subscript indicates that it is the weight associated with the 2nd neuron $a^{(1)}_2$ and the 1st input $x_1$. 

## Forward Pass

### Layer 1 (Hidden Layer)

$$
z_1^{(1)} = (w_{11}^{(1)}x_1 + w_{12}^{(1)}x_2 + b_1^{(1)}) \\
z_2^{(1)} = (w_{21}^{(1)}x_1 + w_{22}^{(1)}x_2 + b_2^{(1)}) \\
z_3^{(1)} = (w_{31}^{(1)}x_1 + w_{32}^{(1)}x_2 + b_3^{(1)}) \\

a_1^{(1)} = \tanh{(z_1^{(1)})} \\
a_2^{(1)} = \tanh{(z_2^{(1)})} \\
a_3^{(1)} = \tanh{(z_3^{(1)})} \\

z_i^{(1)} = b_i^{(1)} + \sum_j w_{ij}^{(1)} x_j \\
a^{(1)}_i = \tanh(z^{(1)}_i)
$$

$$
\mathbf{a}^{(1)} = \tanh(\mathbf{z}^{(1)} + \mathbf{b}^{(1)}) = \tanh(\mathbf{W}^{(1)}\mathbf{x} + \mathbf{b}^{(1)})
$$

where $\mathbf{W}^{(1)} \in \mathbb{R}^{3\times2}$, $\mathbf{b}^{(1)} \in \mathbb{R}^{3\times1}$, $\mathbf{x} \in \mathbb{R}^{2\times1}$, $\mathbf{z}^{(1)} \in \mathbb{R}^{3\times1}$, $\mathbf{a}^{(1)} \in \mathbb{R}^{3\times1}$. The $\tanh$ function is applied element-wise.

### Layer 2 (Output Layer)
$$
\hat{y} = w^{(2)}_{1}a_1^{(1)} + w^{(2)}_{2}a_2^{(1)} + w^{(2)}_{3} a_3^{(1)} + b^{(2)} \\
\hat{y} = b^{(2)} + \sum_j w_j^{(2)} a^{(1)}_j \\
$$

$$
\hat{y} = \begin{bmatrix} w^{(2)}_{1} & w^{(2)}_{2} & w^{(2)}_{3} \end{bmatrix} \begin{bmatrix} a^{(1)}_{1}  \\ a^{(1)}_{2} \\ a^{(1)}_{3} \end{bmatrix} + b^{(2)} = {(\mathbf{w}^{(2)})}^\top \mathbf{a}^{(1)} + b^{(2)}
$$

where ${(\mathbf{w}^{(2)})}^\top \in \mathbb{R}^{1\times3}$, $b^{(2)} \in \mathbb{R}$, $a_1^{(2)} \in \mathbb{R}$. The value $\hat{y}$ is the prediction of the neural network, which is the same as $a_1^{(2)}$.

### Mean Squared Error Loss Function

The mean squared error loss function for a training sample $i$ is defined as

$$
L_t = (\hat{y}_t - y_t)^2
$$

In practice, we feed multiple samples $\mathbf{x}_t$ through the network at the same time, which are a subset of the whole training set. This is known as a mini-batch and is used to estimate the gradient without using the whole dataset. The loss function for the mini-batch with $B$ number of samples is defined as 

$$
L = \frac{1}{B} \sum_{t=1}^B L_t = \frac{1}{B}\sum_{t=1}^B(\hat{y}_t - y_t)^2
$$

The goal is to minimize this loss function, i.e. get the predictions of the NN as close to the true values as possible (without overfitting, but that is for a separate discussion).


## Backward Pass

### Model parameters

In the example neural network, we have 13 parameters in total - 6 weights associated with Layer 1, 3 biases associated with Layer 1, 3 weights associated with Layer 2, and 1 bias associated with Layer 2. We now need to find how to update these weights in order to minimize the loss function. To do so, we will need to find the partial derivatives 

We will denote the weights in Layer 1 as $w_{kl}^{(1)}$, the biases in Layer 1 as $b_k^{(1)}$, the weights in Layer 2 as $w_{k}^{(2)}$ and bias in layer 2 as $b^{(2)}$. We will first go through each one separately, and then compact the equations by using matrix notation.

#### One-sample case

For now, let's stick to the one-example case, where we only have a single sample $\mathbf{x}$ and single output $y$, i.e. we are using a mini-batch size of 1. We need to find the gradient of the loss function with respect to the model parameters. I.e. we need to find:

$$
\frac{\partial L}{\partial w^{(1)}_{kl}}, \frac{\partial L}{\partial w^{(2)}_{k}}, \frac{\partial L}{\partial b^{(1)}_k}, \frac{\partial L}{\partial b^{(2)}}
$$

Note: The subscript here denotes which layer the parameters belong to, it has nothing to do with a second derivative.

To do so, we will use the chain rule:

$$
\frac{\partial L}{\partial w^{(2)}_{k}} = \frac{\partial L}{\partial \hat{y}} \frac{\partial \hat{y}}{\partial w^{(2)}_{k}} \\[0.2cm]
\frac{\partial L}{\partial b^{(2)}} = \frac{\partial L}{\partial \hat{y}} \frac{\partial \hat{y}}{\partial b^{(2)}} \\[0.2cm]
\frac{\partial L}{\partial b^{(1)}_{k}} = \frac{\partial L}{\partial \hat{y}} \frac{\partial \hat{y}}{\partial a^{(1)}_{k}} \frac{\partial a^{(1)}_{k}}{\partial z^{(1)}_{k}} \frac{\partial z^{(1)}_{k}}{\partial b^{(1)}_{k}} \\[0.2cm]
\frac{\partial L}{\partial w^{(1)}_{kl}} = \frac{\partial L}{\partial \hat{y}} \frac{\partial \hat{y}}{\partial a^{(1)}_{k}} \frac{\partial a^{(1)}_{k}}{\partial z^{(1)}_{k}} \frac{\partial z^{(1)}_{k}}{\partial w^{(1)}_{kl}} \\[0.2cm]
$$

Starting from $\frac{\partial L}{\partial \hat{y}}$ we have

$$
\frac{\partial L}{\partial \hat{y}} = \frac{\partial}{{\partial \hat{y}}} (\hat{y} - y)^2 = 2(\hat{y} - y)
$$

Then we have $\frac{\partial \hat{y}}{\partial w_k^{(2)}}$:

$$
\frac{\partial \hat{y}}{\partial w_k^{(2)}} = \frac{\partial}{\partial w_k^{(2)}} \left[ b^{(2)} + \sum_j w_j^{(2)} a^{(1)}_j \right] = 
\sum_j \frac{\partial}{\partial w_k^{(2)}} \left [ w_j^{(2)} a^{(1)}_j \right] = \sum_j \delta_{jk} a^{(1)}_j = a^{(1)}_k
$$

where $\delta_{jk} = 1  \space \text{if} \space j=k \space \text{else} \space 0$.

So we get

$$
\frac{\partial L}{\partial w_k^{(2)}} = 2(\hat{y} - y)a^{(1)}_k
$$

For $\frac{\partial \hat{y}}{\partial b^{(2)}}$ we get:

$$
\frac{\partial \hat{y}}{\partial b^{(2)}} = \frac{\partial}{\partial b^{(2)}} \left[ b^{(2)} + \sum_j w_j^{(2)} a^{(1)}_j \right] = \frac{\partial b^{(2)}}{\partial b^{(2)}} = 1
$$

So we get

$$
\frac{\partial L}{\partial b^{(2)}} = 2(\hat{y} - y)
$$


Moving onto the gradients with respect to the first layer:

$$
\frac{\partial \hat{y}}{\partial a^{(1)}_{k}} = \frac{\partial }{\partial a^{(1)}_{k}} \left[ b^{(2)} + \sum_j w_j^{(2)} a^{(1)}_j \right] = \sum_j \frac{\partial }{\partial a^{(1)}_{k}} \left[ w_j^{(2)} a^{(1)}_j \right] = \sum_j \delta_{jk} w^{(2)}_j = w^{(2)}_k
$$

The derivative of the activation function (depending on the activation function that is used, this derivative will be different):

$$
\frac{\partial a^{(1)}_{k}}{\partial z^{(1)}_{k}} = 1 - \tanh{\left (z^{(1)}_{k} \right )}^2
$$
