# Backpropagation in a Simple Neural Network

## Layer 1

$$
a_1^1 = \tanh{(w_{11}^1x_1 + w_{12}^1x_2 + b_1^1)} \\
a_2^1 = \tanh{(w_{21}^1x_1 + w_{22}^1x_2 + b_2^1)} \\
a_3^1 = \tanh{(w_{31}^1x_1 + w_{32}^1x_2 + b_3^1)} \\
\mathbf{a}^1 = \tanh(\mathbf{W}^1\mathbf{x} + \mathbf{b}^1)
$$

where $\mathbf{W}^1 \in \mathbb{R}^{3\times2}$, $\mathbf{b}^1 \in \mathbb{R}^{3\times1}$, $\mathbf{x} \in \mathbb{R}^{2\times1}$, $\mathbf{a}^1 \in \mathbb{R}^{3\times1}$

## Layer 2 (Output Layer)
$$
\hat{y} = w^2_{11}a_1^1 + w^2_{12}a_2^1 + w^2_{13} a_3^1 + b^2 \\
\hat{y} = \begin{bmatrix} w^2_{11} & w^2_{12} & w^2_{13} \end{bmatrix} \begin{bmatrix} a^1_{1}  \\ a^1_{2} \\ a^1_{3} \end{bmatrix} = {(\mathbf{w}^2)}^\top \mathbf{a}^1
$$

where ${(\mathbf{w}^2)}^\top \in \mathbb{R}^{1\times3}$. The value $\hat{y}$ is the prediction of the neural network.

## Mean Squared Error Loss Function

The mean squared error loss function for this NN example is defined as

$$
MSE = (\hat{y} - y)^2
$$

In practice, we feed multiple samples $\mathbf{x}$ through the network at the same time, which are a subset of the whole training set. This is known as a mini-batch and is used to estimate the gradient without using the whole dataset. The loss function for the mini-batch with $n$ samples is defined as 

$$
MSE_{mini-batch} = \frac{1}{n}\sum_{i=1}^n(\hat{y}_i - y_i)^2
$$
