The subscripts are indices of a matrix or a vector, while superscripts indicate which layer a particular element belongs to. ## Forward Pass

### Layer 1 (Hidden Layer)

$$
z_1^1 = (w_{11}^1x_1 + w_{12}^1x_2 + b_1^1) \\
z_2^1 = (w_{21}^1x_1 + w_{22}^1x_2 + b_2^1) \\
z_3^1 = (w_{31}^1x_1 + w_{32}^1x_2 + b_3^1) \\

a_1^1 = \tanh{(z_1^1)} \\
a_2^1 = \tanh{(z_2^1)} \\
a_3^1 = \tanh{(z_3^1)} \\
\mathbf{a}^1 = \tanh(\mathbf{z}^1 + \mathbf{b}^1) = \tanh(\mathbf{W}^1\mathbf{x} + \mathbf{b}^1)
$$

where $\mathbf{W}^1 \in \mathbb{R}^{3\times2}$, $\mathbf{b}^1 \in \mathbb{R}^{3\times1}$, $\mathbf{x} \in \mathbb{R}^{2\times1}$, $\mathbf{z}^1 \in \mathbb{R}^{3\times1}$, $\mathbf{a}^1 \in \mathbb{R}^{3\times1}$. The $\tanh$ function is applied element-wise.

### Layer 2 (Output Layer)
$$
\hat{y} = a_1^2 = w^2_{11}a_1^1 + w^2_{12}a_2^1 + w^2_{13} a_3^1 + b^2 \\
\hat{y} = a_1^2 = \begin{bmatrix} w^2_{11} & w^2_{12} & w^2_{13} \end{bmatrix} \begin{bmatrix} a^1_{1}  \\ a^1_{2} \\ a^1_{3} \end{bmatrix} + b^2 = {(\mathbf{w}^2)}^\top \mathbf{a}^1 + b^2
$$

where ${(\mathbf{w}^2)}^\top \in \mathbb{R}^{1\times3}$, $b^2 \in \mathbb{R}$, $a_1^2 \in \mathbb{R}$. The value $\hat{y}$ is the prediction of the neural network, which is the same as $a_1^2$.

### Mean Squared Error Loss Function

The mean squared error loss function for a training sample $i$ is defined as

$$
L_i = (\hat{y}_i - y_i)^2
$$

In practice, we feed multiple samples $\mathbf{x}_i$ through the network at the same time, which are a subset of the whole training set. This is known as a mini-batch and is used to estimate the gradient without using the whole dataset. The loss function for the mini-batch with $n$ samples is defined as 

$$
L = \frac{1}{n} \sum_{i=1}^n L_i = \frac{1}{n}\sum_{i=1}^n(\hat{y}_i - y_i)^2
$$

The goal is to minimize this loss function, i.e. get the predictions of the NN as close to the true values as possible (without overfitting the training data, but that is a separate issue).


## Backward Pass

### Model parameters

In the example neural network, we have 13 parameters in total - 6 weights associated with Layer 1, 3 biases associated with Layer 1, 3 weights associated with Layer 2, and 1 bias associated with Layer 2. We now need to find how to update these weights in order to minimize the loss function. To do so, we will need to find the partial derivatives 

We will denote the weights in Layer 1 as $w_{jk}^1$, the biases in Layer 1 as $b_j^1$, the weights in Layer 2 as $w_{jk}^2$ and bias in layer 2 as $b^2$. We will first go through each one separately, and then compact the equations by using matrix notation.

#### One-sample case

For now, let's stick to the one-example case, where we only have a single training input $x$ and output $y$. We need to find the gradient of the loss function with respect to the model parameters. I.e. we need to find:

$$
\frac{\partial L}{\partial w^1_{jk}}, \frac{\partial L}{\partial w^2_{jk}}, \frac{\partial L}{\partial b^1_j}, \frac{\partial L}{\partial b^2}
$$

Note: The subscript here denotes which layer the parameters belong to, it has nothing to do with a second derivative.

To do so, we will use the chain rule:

$$
\frac{\partial L}{\partial w^1_{jk}} = \frac{\partial L}{\partial \hat{y}} \frac{\partial \hat{y}}{\partial a^1_{m}} \frac{\partial a^1_{m}}{\partial z^1_{m}} \frac{\partial z^1_{m}}{\partial w^1_{jk}} \\[0.2cm]
\frac{\partial L}{\partial w^2_{jk}} = \frac{\partial L}{\partial \hat{y}} \frac{\partial \hat{y}}{\partial w^2_{jk}} \\[0.2cm] 
\frac{\partial L}{\partial b^1_{j}} = \frac{\partial L}{\partial \hat{y}} \frac{\partial \hat{y}}{\partial a^1_{m}} \frac{\partial a^1_{m}}{\partial z^1_{m}} \frac{\partial z^1_{m}}{\partial b^1_{m}} \\[0.2cm]
\frac{\partial L}{\partial b^2} = \frac{\partial L}{\partial \hat{y}} \frac{\partial \hat{y}}{\partial b^2}
$$
